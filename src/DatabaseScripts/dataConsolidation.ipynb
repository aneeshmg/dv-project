```python
import csv
import pymongo
from pymongo import MongoClient

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["dv"]

#mycol1 = mydb.impact_scores
mycol2 = mydb.businesses
#mycol3 = mydb.review_with_stars
rows = []
i = 0

with open('utput_2.csv', 'r', encoding="utf8") as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    csv_output_file = open('last3.csv', mode='w', encoding="utf8")
    fieldnames = ['user_id', 'business_id', 'review_id', 'review_text', 'topic', 'sentiment_score', 'date','stars','impact_score', 'business_name']
    writer = csv.DictWriter(csv_output_file, fieldnames=fieldnames)
    writer.writeheader()
    for row in readCSV:
        #print(row[0])
        try:
            #score = mycol1.find_one({'user_id':str(row[0])},{'impact_score' : 1})['impact_score']
            bizname = mycol2.find_one({'business_id':str(row[1])},{'name' : 1})['name']
            #starval = mycol3.find_one({'review_id':str(row[2])},{'stars' : 1})['stars']

        except:
            #score = 0
            bizname =""
            #starval = 0
            
        row_o = {
            'user_id':row[0],
            'business_id':row[1],
            'review_id':row[2],
            'review_text':row[3],
            'topic':row[4],
            'sentiment_score':row[5],
            'date':row[6],
            'stars':row[7],
            'impact_score':row[8],
            'business_name':bizname
        }
        i = i + 1
        rows.append(row_o)
        if len(rows) > 1000:
            print(i)
            writer.writerows(rows)
            rows = []
```
